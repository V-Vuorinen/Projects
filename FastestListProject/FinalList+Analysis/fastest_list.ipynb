```python
import pandas as pd
import sqlalchemy
import getpass
```

```python
file_path = r'C:\Users\vaino\Desktop\Python_jupyter\Fastest_List.csv'
```

```python
# - Crate variables to hold credentials for the database connection
# - Create engine to thee database
# - HINT! password is 12345 :)

table_name = 'fastest_list'
db_user = 'postgres'
db_password = getpass.getpass("Enter Password")
db_host = 'localhost'
db_port = '5432'
db_name = 'postgres'

credentials = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
engine = sqlalchemy.create_engine(credentials)
```

```python
query_cars = 'SELECT * FROM fastest_cars;'
query_animals = 'SELECT * FROM fastest_animals;'
query_tennis = 'SELECT * FROM fastest_tennis;'
```

```python
df_cars = pd.read_sql_query(query_cars, engine)
df_animals = pd.read_sql_query(query_animals, engine)
df_tennis = pd.read_sql_query(query_tennis, engine)
engine.dispose()
```

```python
# - Look at the number of columns and rows of each dataframe
# - We want to take the top 20 ranked of each dataframe

print(df_cars.shape)
print(df_animals.shape)
print(df_tennis.shape)
```

```python
df_cars = df_cars.head(20)
df_animals = df_animals.head(20)
df_tennis = df_tennis.head(20)
```

```python
# - Take a look at each dataframe and get them ready to concat
df_cars.head()
```

```python
df_animals.head()
```

```python
df_tennis.head()
```

```python
# - df_cars
# - We wil merge the columns 'Make' and 'Model' into 'Name'
# - We will also create new column 'Specification' including the specification of the species (car)

df_cars['Name'] = df_cars['Make'] + ' ' + df_cars['Model']
df_cars = df_cars.drop(['Make', 'Model'], axis=1)
df_cars = df_cars.rename(columns={'Top speed':'Top Speed'})

df_cars['Specification'] = 'Car'
df_cars.head(2)
```

```python
# - df_animals
# - We will drop column 'Notes'
# - Change columns 'Animal' and 'Maximum Speed' into 'Name' and 'Top Speed'
# - Finaly we will change column 'Class' into 'Specification'

df_animals = df_animals.drop('Notes', axis=1)
df_animals = df_animals.rename(columns={'Maximum Speed': 'Top Speed', 'Animal': 'Name', 'Class': 'Specification'})
df_animals.head(2)
```

```python
# - df_tennis
# - Drop column 'Country'
# - Rename columns 'Player' and 'Speed' into 'Name' and 'Top Speed
# - Add column 'Specification' and add 'Tennis ball' to all rows

df_tennis = df_tennis.drop('Country', axis=1)
df_tennis = df_tennis.rename(columns={'Player': 'Name', 'Speed': 'Top Speed'})
df_tennis['Specification'] = 'Tennis ball'
df_tennis.head()
```

```python
# - Merge the dataframes

df = pd.concat([df_cars, df_animals, df_tennis], ignore_index=True)
```

```python
# - Check the dataframe

df = df.sort_values(by='Top Speed', ascending=False)
df.reset_index(drop=True, inplace=True)
df
```

```python
df['Top Speed'].unique()
```

```python
# - Replace values with '-' and use number after it

df.loc[df['Top Speed'] == '240–320 km/h', 'Top Speed'] = '320 km/h'
df.loc[df['Top Speed'] == '109.4–120.7 km/h', 'Top Speed'] = '120.7 km/h'
```

```python
# - Slice the column 'Top Speed' to show only the number value of Top Speed
# - Create a blank column containing 'km/h'
# - Change order of columns

df['Top Speed'] = df['Top Speed'].str.extract(r'(\d+)').astype(float).astype('Int64')
df[' '] = 'km/h'
df = df.sort_values(by='Top Speed', ascending=False)

column_order = ['Rank', 'Name', 'Specification', 'Top Speed', ' ']
df = df[column_order]
df
```

```python
# - Change specification names

df.loc[df['Specification'] == 'Flight-diving', 'Specification'] = 'Bird'
df.loc[df['Specification'] == 'Flight', 'Specification'] = 'Bird'
df.loc[df['Specification'] == 'Flight-swimming', 'Specification'] = 'Fish'
df.loc[df['Specification'] == 'Swimming', 'Specification'] = 'Fish'
df.loc[df['Specification'] == 'Land', 'Specification'] = 'Land species'
```

```python
df
```

```python
# - Write the dataframe into the database using the engine
# - Use the credentials created earlier and remember to dispose the engine

df.to_sql(table_name, engine, if_exists='replace', index=False)

engine.dispose()
```

```python
# - Write thee dataframe into a XML file located in the path created earlier

df.to_csv(file_path)
```
