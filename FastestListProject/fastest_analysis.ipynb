```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
import sqlalchemy
import getpass
```

```python
# - Crate variables to hold credentials for the database connection
# - Create engine to thee database
# - HINT! password is 12345 :)

table_name = 'fastest_list'
db_user = 'postgres'
db_password = getpass.getpass("Enter Password")
db_host = 'localhost'
db_port = '5432'
db_name = 'postgres'

credentials = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
engine = sqlalchemy.create_engine(credentials)
```

```python
# - Query the list from the database into a dataframe

query_list = 'SELECT * FROM fastest_list;'

df = pd.read_sql_query(query_list, engine)
engine.dispose()
```

```python
# - Take look at the dataframe 

df
```

```python
# - Update the ranks and edit them into the same format

df['Rank'] = df['Rank'].astype(str).str.rstrip('.')
df['Rank'] = range(1, len(df) + 1)
df
```

```python
# - This dataset is very small and we know it has no null or NaN values
# - Because of this we will not use anymore time in formatting the list

df.shape
```

```python
# - Lets start by plotting a barplot from sns and try other plotting types
# - Here we can see that cars had a significantly high average speed compared to other //
# - specifications. The distribution inside the cars specification was also was very low, //
# - when for example the values within birds had a more vibrant distribution.

plt.figure(figsize=(12, 6))
sns.barplot(x='Specification', y='Top Speed', data=df)
plt.title('Average Top Speed by Specification')
plt.show()
```

```python
# - Lets try to understand better the distribution of values per specification type //
# - we will use a boxplot for this. We better understand the significanse of the distribution //
# - within birds. We also see that the Tennis ball hits, cars and fish had low variance //
# - within the categories. 

plt.figure(figsize=(12, 6))
sns.boxplot(x='Specification', y='Top Speed', data=df)
plt.title('Top Speed Distribution by Specification')
plt.show()
```

```python
# - Scatter plot gives a more detailed view of the same theory

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Rank', y='Top Speed', hue='Specification', data=df)
plt.title('Top Speed vs Rank')
plt.show()
```

```python
# - Here we can understand the overall population of each specification type in the list. 
# - The results would be different, if we wouldn't have limitted the initial lists //
# - into top 20 of each list (fastest_cars, fastest_animals, fastest_tennis).

plt.figure(figsize=(10, 6))
sns.countplot(x='Specification', data=df)
plt.title('Count of Entries by Specification')
plt.show()
```

```python
# - The facet grid is a great tool to view variance within specification categories, since //
# - the scatterplots can sometimes be hard to read percicely

g = sns.FacetGrid(df, col="Specification", height=4)
g.map(plt.scatter, "Rank", "Top Speed")
plt.show()
```
