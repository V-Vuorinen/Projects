```python
import requests
import sqlalchemy
import getpass
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# - Define target URL and file path for future use
# - Create dataframe for future use

url = 'https://tenniscompanion.org/fastest-tennis-serves/'
xml_path = r'C:\Users\vaino\Desktop\Python_jupyter\Fastest_Tennis_Serves.xml'
df = pd.DataFrame(columns=["Rank", "Country", "Player", "Speed"])
```

```python
# - Crate variables to hold credentials for the database connection
# - HINT! password is 12345 :)

table_name = 'fastest_tennis'
db_user = 'postgres'
db_password = getpass.getpass("Enter Password")
db_host = 'localhost'
db_port = '5432'
db_name = 'postgres'
```

```python
# - Load essentials from webpage

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('tbody')
rows = table.find_all('tr')[1:]
```

```python
# - Extract the data
# - Write the data in the dataframe created earlier

for row in rows:
        col = row.find_all('td')
        if len(col) != 0:
            data_dict = {"Rank": col[0].get_text(strip=True),
                         "Country": col[1].get_text(strip=True),
                         "Player": col[2].get_text(strip=True),
                         "Speed": col[3].get_text(strip=True)}
            df1 = pd.DataFrame(data_dict, index=[0])
            df = pd.concat([df, df1], ignore_index=True)
        else:
            break
```

```python
# - Check dataframe

df
```

```python
# - Get rid of mph value in brackets

df["Speed"] = df["Speed"].str.split().str[:2]
```

```python
# - Use lambda function to format the Speed values

df['Speed'] = df['Speed'].apply(lambda x: [x[0].replace('[', ''), x[1].replace(']', '')])
df['Speed'] = df['Speed'].apply(lambda x: ' '.join(x))
```

```python
df
```

```python
# - Create engine for connecting into the database
# - Use the credentials created earlier and remember to dispose the engine

credentials = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
engine = sqlalchemy.create_engine(credentials)

df.to_sql(table_name, engine, if_exists='replace', index=False)

engine.dispose()
```

```python
# - Write thee dataframe into a XML file located in the path created earlier

df.to_xml(xml_path)
```
